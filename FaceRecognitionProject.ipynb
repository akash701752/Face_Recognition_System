Importing Libraries

```python
import cv2
import numpy as np
import face_recognition
```

```python
from google.colab.patches import cv2_imshow
```

```python
imgVirat = face_recognition.load_image_file('/content/drive/MyDrive/Colab Notebooks/FaceRecognition/ImageBasicss/VirtaKohli.jpg')
imgVirat = cv2.cvtColor(imgVirat,cv2.COLOR_BGR2RGB) 

#cv2.imshow() does not work on google colab
cv2_imshow(imgVirat)
cv2.waitKey(0)
```

```python
#Testing Image
imgTest = face_recognition.load_image_file('/content/drive/MyDrive/Colab Notebooks/FaceRecognition/ImageBasicss/ViratTest.jpg')
imgTest = cv2.cvtColor(imgTest,cv2.COLOR_BGR2RGB) 

cv2_imshow(imgTest)
```

```python
#Making Rectangle
faceLoc = face_recognition.face_locations(imgVirat)[0]
encodeVirat = face_recognition.face_encodings(imgVirat)[0]

print(faceLoc)
cv2.rectangle(imgVirat,(faceLoc[3],faceLoc[0]),(faceLoc[1],faceLoc[2]),(255,0,255),2)
cv2_imshow(imgVirat)
cv2_imshow(imgTest)
```

```python
faceLocTest = face_recognition.face_locations(imgTest)[0]
encodeViratTest = face_recognition.face_encodings(imgTest)[0]
cv2.rectangle(imgTest,(faceLocTest[3],faceLocTest[0]),(faceLocTest[1],faceLocTest[2]),(255,0,255),2)
cv2_imshow(imgVirat)
cv2_imshow(imgTest)
```

```python
#comparing faces
results = face_recognition.compare_faces([encodeVirat],encodeViratTest)

#finding Distance
faceDis = face_recognition.face_distance([encodeVirat],encodeViratTest)

print(results,faceDis)
```

```python
#giving Name to face
cv2.putText(imgTest,f'{results} {round(faceDis[0],2)}',(50,50),cv2.FONT_HERSHEY_COMPLEX,1,(0,0,255),2)

cv2_imshow(imgVirat)
cv2_imshow(imgTest)
```
